```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = []
with open('tag_data.csv' ,'r') as f:
    lines = f.readlines()
    for line in lines:
        line = line.strip()
        line = line.split(',')
        line = [float(x) for x in line]
        data.append(line)
data = np.array(data)

actuals = data.T[0]
estimated = data.T[1]

grouped_data = {actual:[] for actual in actuals}
for actual, estimate in data:
    grouped_data[actual].append(estimate)
    
plt.plot(actuals, estimated, 'o')
plt.plot(actuals, actuals)
plt.show()
plt.plot(actuals - estimated, 'o')
```

```python
data_means = np.array([[actual, np.mean(grouped_data[actual])] for actual in grouped_data.keys()])
plt.plot(data_means.T[0], data_means.T[1], 'o')
plt.plot(data_means.T[0], data_means.T[0])
plt.show()
plt.plot(data_means.T[0] - data_means.T[1], 'o')
print(data_means.T[0] - data_means.T[1])
```

```python
data_std = np.array([[actual, np.std(grouped_data[actual])] for actual in grouped_data.keys()])
plt.plot(data_std.T[0], data_std.T[1], 'o')
plt.show()
```

```python
actuals = data.T[0]
estimated = data.T[1]
counts, bins = np.histogram(estimated - actuals, bins=50)
plt.stairs(counts, bins)
plt.show()

for actual in grouped_data.keys():
    counts, bins = np.histogram(grouped_data[actual] - actual, bins=10)
    plt.stairs(counts, bins)
    plt.show()
```
